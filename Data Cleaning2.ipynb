```python
clean_df = df[['Age','Married', 'Number of Dependents',
       'City', 'Number of Referrals',
       'Tenure in Months', 'Offer', 'Phone Service',
       'Avg Monthly Long Distance Charges', 'Multiple Lines',
       'Internet Service', 'Internet Type', 'Online Security', 'Online Backup', 'Device Protection Plan', 'Contract', 'Payment Method', 'Monthly Charge', 'Total Charges', 'Total Refunds',
       'Total Extra Data Charges', 'Total Long Distance Charges',
       'Total Revenue', 'Customer Status']]
```

```python
clean_df[clean_df['Multiple Lines'].isnull()]
# Null values in Avg Long distance in Avg monthly charges and multiple lines is consistent
```

```python
from sklearn.impute import KNNImputer

clean_df2 = clean_df
clean_df2['Multiple Lines'] = clean_df2['Multiple Lines'].replace({'No':0, 'Yes':1})

# Select the columns for imputation
imputer = KNNImputer(n_neighbors=5)
clean_df2[['Avg Monthly Long Distance Charges','Multiple Lines']] = imputer.fit_transform(clean_df2[['Avg Monthly Long Distance Charges','Multiple Lines']])
```

```python
clean_df2
```

```python
clean_df2['Multiple Lines'] = clean_df2['Multiple Lines'].astype('int')
clean_df2['Multiple Lines'].replace({0:'No', 1:'Yes'}, inplace = True)
clean_df2
```

```python
# understand null values pattern for Internet type
clean_df2[clean_df2['Internet Type'].isnull()]
```

Here the pattern is where Internet Type is No, Online Security, Online Backup. and Device Protection plan is not recorded

```python
# Filled the null values with 'Not Specified'
clean_df2[['Internet Type',	'Online Security', 'Online Backup',	'Device Protection Plan']] = clean_df2[['Internet Type','Online Security', 'Online Backup',	'Device Protection Plan']].fillna('Not Specified')
```
